```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page_list=[]
for page in range(1,101):
    
    a = "https://esf.fang.com/house-a015277/i3{}".format(page)
    tongguo = a +'/?rfss=2-bc0b344d1eabc0c05f-38'
    page_list.append(tongguo)

page_list
```

```python
url_list=[]
root_url = 'https://esf.fang.com'
for page in page_list:
    response = requests.get(page)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find_all('h4', class_='clearfix')
    for result in results:
        relative_url = result.find('a').get('href')
        url_combine = root_url+relative_url
    
        url_list.append(url_combine)
```
